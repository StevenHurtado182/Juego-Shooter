```python
import pygame
import random

# Inicializar pygame
pygame.init()

# Definir dimensiones de la pantalla
WIDTH, HEIGHT = 800, 600
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("CODCITO")

# Colores
WHITE = (255, 255, 255)
RED = (255, 0, 0)
BLUE = (0, 0, 255)

# velocidad de fotogramas
clock = pygame.time.Clock()

# imágenes
inave = pygame.image.load("nave.png")  
ienemigo = pygame.image.load("enemigo.png") 

# Clases
class Jugador(pygame.sprite.Sprite):
    def __init__(self):
        super().__init__()
        self.image = pygame.transform.scale(inave, (50, 50))
        self.rect = self.image.get_rect()
        self.rect.centerx = WIDTH // 2
        self.rect.bottom = HEIGHT - 10
        self.speed_x = 0

    def update(self):
        self.rect.x += self.speed_x
        if self.rect.left < 0:
            self.rect.left = 0
        if self.rect.right > WIDTH:
            self.rect.right = WIDTH

    def disparos(self):
        balas = Balas(self.rect.centerx, self.rect.top)
        all_sprites.add(balas)
        sbalas.add(balas)

class Enemigo(pygame.sprite.Sprite):
    def __init__(self):
        super().__init__()
        self.image = pygame.transform.scale(ienemigo, (50, 50))
        self.rect = self.image.get_rect()
        self.rect.x = random.randint(0, WIDTH - self.rect.width)
        self.rect.y = random.randint(-100, -40)
        self.speed_y = random.randint(1, 5)

    def update(self):
        self.rect.y += self.speed_y
        if self.rect.top > HEIGHT:
            self.rect.x = random.randint(0, WIDTH - self.rect.width)
            self.rect.y = random.randint(-100, -40)
            self.speed_y = random.randint(1, 5)
            return True
        return False

class Balas(pygame.sprite.Sprite):
    def __init__(self, x, y):
        super().__init__()
        self.image = pygame.Surface((5, 15))
        self.image.fill(RED)
        self.rect = self.image.get_rect()
        self.rect.centerx = x
        self.rect.bottom = y
        self.speed_y = -10

    def update(self):
        self.rect.y += self.speed_y
        if self.rect.bottom < 0:
            self.kill()

# Grupos de sprites
all_sprites = pygame.sprite.Group()
senemigo = pygame.sprite.Group()
sbalas = pygame.sprite.Group()

# Crear jugador
jugador = Jugador()
all_sprites.add(jugador)

# Crear enemigos
for _ in range(2):
    enemigo = Enemigo()
    all_sprites.add(enemigo)
    senemigo.add(enemigo)

enemigos_pasados = 0

# Bucle principal del juego
running = True
while running:
    clock.tick(30)

    # Procesar eventos
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        elif event.type == pygame.KEYDOWN:
            if event.key == pygame.K_LEFT:
                jugador.speed_x = -10
            elif event.key == pygame.K_RIGHT:
                jugador.speed_x = 10
            elif event.key == pygame.K_SPACE:
                jugador.disparos()
        elif event.type == pygame.KEYUP:
            if event.key == pygame.K_LEFT or event.key == pygame.K_RIGHT:
                jugador.speed_x = 0

    # Actualizar
    all_sprites.update()
    
    # Verificar si algún enemigo pasa la parte inferior
    for enemigo in senemigo:
        if enemigo.update():  # Si el enemigo ha pasado la pantalla
            enemigos_pasados += 1
            if enemigos_pasados >= 3:
                print("¡Perdiste!")
                running = False  # Termina el juego si 3 enemigos han pasado
    
    # Colisiones entre balas y enemigos
    impacto = pygame.sprite.groupcollide(senemigo, sbalas, True, True)
    for i in impacto:
        enemigo = Enemigo()
        all_sprites.add(enemigo)
        senemigo.add(enemigo)

    # Dibujar / Renderizar
    screen.fill(WHITE)
    all_sprites.draw(screen)

    # Mostrar en pantalla
    pygame.display.flip()

pygame.quit()
```

```python
import pygame
import random

# Inicializar pygame
pygame.init()

# Definir dimensiones de la pantalla
WIDTH, HEIGHT = 800, 600
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("CODCITO")

# Colores
WHITE = (255, 255, 255)
RED = (255, 0, 0)
BLUE = (0, 0, 255)
BLACK = (0, 0, 0)
GREEN = (0, 255, 0)

# velocidad de fotogramas
clock = pygame.time.Clock()

# imágenes
inave = pygame.image.load("nave.png")
ienemigo = pygame.image.load("enemigo.png")

# Definir fuente para el texto
font_large = pygame.font.SysFont(None, 74)
font_small = pygame.font.SysFont(None, 36)

# Clases
class Jugador(pygame.sprite.Sprite):
    def __init__(self):
        super().__init__()
        self.image = pygame.transform.scale(inave, (50, 50))
        self.rect = self.image.get_rect()
        self.rect.centerx = WIDTH // 2
        self.rect.bottom = HEIGHT - 10
        self.speed_x = 0

    def update(self):
        self.rect.x += self.speed_x
        if self.rect.left < 0:
            self.rect.left = 0
        if self.rect.right > WIDTH:
            self.rect.right = WIDTH

    def disparos(self):
        balas = Balas(self.rect.centerx, self.rect.top)
        all_sprites.add(balas)
        sbalas.add(balas)

class Enemigo(pygame.sprite.Sprite):
    def __init__(self):
        super().__init__()
        self.image = pygame.transform.scale(ienemigo, (50, 50))
        self.rect = self.image.get_rect()
        self.rect.x = random.randint(0, WIDTH - self.rect.width)
        self.rect.y = random.randint(-100, -40)
        self.speed_y = random.randint(1, 5)

    def update(self):
        self.rect.y += self.speed_y
        if self.rect.top > HEIGHT:
            self.rect.x = random.randint(0, WIDTH - self.rect.width)
            self.rect.y = random.randint(-100, -40)
            self.speed_y = random.randint(1, 5)
            return True
        return False

class Balas(pygame.sprite.Sprite):
    def __init__(self, x, y):
        super().__init__()
        self.image = pygame.Surface((5, 15))
        self.image.fill(RED)
        self.rect = self.image.get_rect()
        self.rect.centerx = x
        self.rect.bottom = y
        self.speed_y = -10

    def update(self):
        self.rect.y += self.speed_y
        if self.rect.bottom < 0:
            self.kill()

# Grupos de sprites
all_sprites = pygame.sprite.Group()
senemigo = pygame.sprite.Group()
sbalas = pygame.sprite.Group()

# Crear jugador
jugador = Jugador()
all_sprites.add(jugador)

# Crear enemigos
for _ in range(2):
    enemigo = Enemigo()
    all_sprites.add(enemigo)
    senemigo.add(enemigo)

# Contador de enemigos que pasan
enemigos_pasados = 0

# Bucle principal del juego
running = True
game_over = False

while running:
    clock.tick(30)

    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        elif event.type == pygame.KEYDOWN:
            if event.key == pygame.K_LEFT:
                jugador.speed_x = -10
            elif event.key == pygame.K_RIGHT:
                jugador.speed_x = 10
            elif event.key == pygame.K_SPACE:
                jugador.disparos()
        elif event.type == pygame.KEYUP:
            if event.key == pygame.K_LEFT or event.key == pygame.K_RIGHT:
                jugador.speed_x = 0
        elif event.type == pygame.MOUSEBUTTONDOWN and game_over:
            mouse_pos = pygame.mouse.get_pos()
            if play_again_button.collidepoint(mouse_pos):
                # Resetear el juego
                game_over = False
                enemigos_pasados = 0
                all_sprites.empty()
                senemigo.empty()
                sbalas.empty()
                jugador = Jugador()
                all_sprites.add(jugador)
                for _ in range(2):
                    enemigo = Enemigo()
                    all_sprites.add(enemigo)
                    senemigo.add(enemigo)
            elif quit_button.collidepoint(mouse_pos):
                running = False
    

    if not game_over:
        # Actualizar
        all_sprites.update()

        # Verificar si algún enemigo pasa la parte inferior
        for enemigo in senemigo:
            if enemigo.update():  # Si el enemigo ha pasado la pantalla
                enemigos_pasados += 1
                if enemigos_pasados >= 3:
                    game_over = True

        # Colisiones entre balas y enemigos
        impacto = pygame.sprite.groupcollide(senemigo, sbalas, True, True)
        for i in impacto:
            enemigo = Enemigo()
            all_sprites.add(enemigo)
            senemigo.add(enemigo)

        # Dibujar / Renderizar
        screen.fill(WHITE)
        all_sprites.draw(screen)
    else:
        # Mostrar pantalla de game over
        screen.fill(WHITE)
        game_over_text = font_large.render("¡Perdiste!", True, RED)
        screen.blit(game_over_text, (WIDTH // 2 - game_over_text.get_width() // 2, HEIGHT // 4))

        # Crear botones de "Volver a jugar" y "Salir"
        play_again_text = font_small.render("Volver a jugar", True, WHITE)
        quit_text = font_small.render("Salir", True, WHITE)
        
        play_again_button = pygame.Rect(WIDTH // 2 - 100, HEIGHT // 2, 200, 50)
        quit_button = pygame.Rect(WIDTH // 2 - 100, HEIGHT // 2 + 60, 200, 50)
        
        pygame.draw.rect(screen, GREEN, play_again_button)
        pygame.draw.rect(screen, BLUE, quit_button)
        
        screen.blit(play_again_text, (play_again_button.centerx - play_again_text.get_width() // 2, play_again_button.centery - play_again_text.get_height() // 2))
        screen.blit(quit_text, (quit_button.centerx - quit_text.get_width() // 2, quit_button.centery - quit_text.get_height() // 2))

    pygame.display.flip()

pygame.quit()
```
